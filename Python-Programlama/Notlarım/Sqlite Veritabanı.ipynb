Siz de bir tane Şarkı projesi geliştirmeye çalışın.

                     Örnek özellikler;

                     1. Şarkı İsmi 
                     2. Sanatçı
                     3. Albüm
                     4. Prodüksiyon Şirketi
                     5. Şarkı Süresi

                     Örnek Metodlar;

                     1. Veritabanındaki Toplam Şarkı Süresini Hesaplayan Metod
                     2. Şarkı Ekle
                     3. Şarkı Sil

```python
import sqlite3

class Şarkı():
    def __init__(self,isim,sanatçı,albüm,prodüksiyon_şirketi,şarkı_süresi):
        self.isim = isim
        self.sanatçı = sanatçı
        self.albüm = albüm
        self.prodüksiyon_şirketi = prodüksiyon_şirketi
        self.şarkı_süresi = şarkı_süresi
    def __str__(self):
        return "Şarkı İsmi: {}\nSanatçı: {}\nAlbüm: {}\nProdüksiyon Şirketi: {}\nŞarkı Süresi: {}".format(self.isim,self.sanatçı,self.albüm,self.prodüksiyon_şirketi,self.şarkı_süresi)

class Radyo():
    def __init__(self):
        self.con = sqlite3.connect("radyo.db")
        self.cur = self.con.cursor()
        self.cur.execute("create table if not exists Şarkılar(İsim text, Sanatçı text, Albüm text, Prodüksiyon_Şirketi text, Şarkı_Süresi int)")
        self.con.commit()
    def şarkı_ekle(self,şarkı):
        self.cur.execute("insert into Şarkılar values(?,?,?,?,?)",(şarkı.isim,şarkı.sanatçı,şarkı.albüm,şarkı.prodüksiyon_şirketi,şarkı.şarkı_süresi))
        self.con.commit()
    def şarkı_listesi(self):
        self.cur.execute("select * from Şarkılar")
        liste = self.cur.fetchall()
        for i in liste:
            şarkı = Şarkı(i[0],i[1],i[2],i[3],i[4])
            print(şarkı)
    def şarkı_sil(self,isim):
        self.cur.execute("select * from Şarkılar where İsim = ?",(isim,))
        liste = self.cur.fetchall()
        if len(liste) == 0:
            print("Bu şarkı listemizde yok.")
        else:
            self.cur.execute("delete from Şarkılar where İsim = ?",(isim,))
    def toplam_şarkı_süresi(self):
        toplam = 0
        self.cur.execute("select * from Şarkılar")
        liste = self.cur.fetchall()
        for i in liste:
            toplam += i[4]
        return toplam

radyo = Radyo()

while True:
    işlem = input("İşlem seçiniz:")
    if işlem == "q":
        print("Program Kapanıyor")
        break
    elif işlem == "1":
        print("Şarkı Listemiz:")
        radyo.şarkı_listesi()
    elif işlem == "2":
        isim = input("Şarkı ismi:")
        sanatçı = input("Sanatçı:")
        albüm = input("Albüm:")
        prodüksiyon_şirketi = input("Prodüksiyon Şirketi:")
        şarkı_süresi = int(input("Şarkı Süresi:"))
        
        şarkı = Şarkı(isim,sanatçı,albüm,prodüksiyon_şirketi,şarkı_süresi)
        
        radyo.şarkı_ekle(şarkı)
    elif işlem == "3":
        isim = input("Silmek istediğiniz şarkı ismi:")
        radyo.şarkı_sil(isim)
    elif işlem == "4":
        print("Toplam Şarkı Süresi: {}".format(radyo.toplam_şarkı_süresi()))
```
